```python
# rank:
Student Name    Total marks   Rank(based on max val)     Rank(based on min val)
A               100           4                          1
B               200           3                          2
C               300           2                          3
D               400           1                          4
```

```python
import numpy as np
np.random.seed(10)
data =pd.DataFrame({'name':['a','b','c','d','e'],'balance':np.random.randint(1,1000,5),'defaults':np.random.randint(1,15,5)})
data
```

```python
data['rank_balance'] = data['balance'].rank()  #default ranking is ascending
data
```

```python
data['rank_balance'] = data['balance'].rank(ascending=False)  #default ranking is ascending
data
```

```python
data['rank_default'] = data['defaults'].rank()
data
```

```python
#if two rows(elements) are equal then an average of values is assigned to those elements
```

```python
data =pd.DataFrame({'name':['a','b','c','d','e'],'marks':[300,400,200,200,350]})
data['rank_average'] = data['marks'].rank(ascending=True)
data
```

```python
data['rank_min'] = data['marks'].rank(ascending=False,method='min')
data
```

```python
data['rank_max'] = data['marks'].rank(ascending=False,method='max')
data
```

```python
data['rank_dense'] = data['marks'].rank(ascending=False,method='dense')
data
```
